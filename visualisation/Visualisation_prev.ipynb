# CLASSES


```python
#Nos classes
NosClasses = dataset.iloc[:, 7:15]
print(len(NosClasses))
totals = NosClasses.sum()
len(totals)
totals
```

```python
extended_labels = ['Normal', 'Diabetes', 'Glaucoma', 'Cataract', 'Age related Macular Degeneration', 'Hypertension', 'Pathological Myopia', 'Other abnormalities']
plt.pie(totals, labels=extended_labels, startangle=90)
plt.title('Répartition globale de nos classes').set_fontsize(20)
plt.show()
```

# AGE


```python
fig, (ax1) = plt.subplots(1, 1, figsize=(15,5))
sns.countplot(ax=ax1, x="Patient Age", data=dataset)
ax1.set_title("distribution d'Age")
plt.show()
```

# SEX

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(20,5))
sns.countplot(ax=ax1, x="Patient Sex", data=dataset)
ax1.set_title("Distribution de Sex ").set_fontsize(20)
plt.show()
```

```python
a=["N","D","G","C","A","H","M","O"]
```

```python
sns.catplot(x="Patient Sex", y="Patient Age", hue='labels',data=dataset)
plt.title("Relation classe et sex et age ").set_fontsize(20)
plt.show()
```

```python
sns.set_style("darkgrid")
fig= plt.subplots(figsize=(20,10))
sns.countplot(x="Patient Age", hue='M', data=dataset)
plt.title("Relation classe et sex ").set_fontsize(20)
plt.show()
```

```python
sns.set_style("darkgrid")
fig= plt.subplots(figsize=(20,10))
sns.countplot(x="Patient Sex", hue='labels', data=dataset)
plt.title("Relation classe et sex ").set_fontsize(20)
plt.show()
```

```python
for e in a:
    sns.set_style("darkgrid")
    fig= plt.subplots(figsize=(20,5))
    sns.countplot(x="Patient Age", hue=e, data=dataset)
    plt.title("Diagnostics " + e +" regroupés par ").set_fontsize(20)
    plt.show()
```
